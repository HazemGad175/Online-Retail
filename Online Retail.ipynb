```python
df = df.drop_duplicates().reset_index()
df
```

```python
pd.set_option('display.max.rows', 60)
top_seller = df['Description'].value_counts().head(10).reset_index()
print('top 10 '  , top_seller)
```

```python
df['category'] = df['Description'].apply(lambda x:
    'Kitchen' if any( word in str(x).upper() for word in ['CAKE', 'BAKING', 'TEA', 'JAM', 'CUP', 'RECIPE', 'PANTRY', 'MUG', 'PLATE', 'BOWL', 'TRAY', 'TOWEL', 'COASTER'])
    else 'Kids' if any( word in str(x).upper() for word in ['CHILDREN', 'SPACEBOY', 'DOLLY', 'CIRCUS', 'NAP', 'TOY'])
    else 'Lighting' if any(word in str(x).upper() for word in ['LIGHT', 'LANTERN', 'CANDLE', 'T-LIGHT'])
    else 'Bags' if any(word in str(x).upper() for word in ['BAG', 'SHOPPER', 'LUNCH', 'JUMBO', 'PURSE', 'PENCIL CASE'])
    else 'Decor' if any(word in str(x).upper() for word in ['HEART', 'FRAME', 'HANGING', 'CHALKBOARD', 'SIGN', 'WALL ART'])
    else 'Craft' if any(word in str(x).upper() for word in ['KNITTED', 'WOODEN', 'CRAFT', 'MAKING', 'PAINT', 'STITCH'])
    else 'Homeware' if any(word in str(x).upper() for word in ['DOORMAT', 'CUSHION', 'COVER', 'VASE', 'CLOCK', 'STORAGE'])
    else 'Gifts' if any(word in str(x).upper() for word in ['GIFT', 'PRESENT', 'WRAP', 'RIBBON', 'CARD', 'TAGS', 'TAPE'])
    else 'Seasonal' if any( word in str(x).upper() for word in ['CHRISTMAS', 'XMAS', 'EASTER', 'HALLOWEEN', 'VALENTINE', 'ADVENT'])
    else 'Tools' if any( word in str( x).upper() for word in ['TOOL', 'SCISSORS', 'TAPE', 'KNEELING', 'PAD', 'TOOLSET'])
    else 'Garden' if any(word in str(x ).upper() for word in ['GARDEN', 'PLANT', 'POT', 'FLOWER', 'SEEDS'])
    else 'Stationery' if any(word in str( x).upper() for word in ['NOTEBOOK', 'PEN', 'PENCIL', 'ERASER', 'SHARPENER', 'RULER', 'DIARY'])
    else 'Other')
df
```

```python
df['year']= df['InvoiceDate'].dt.year
df['month']= df['InvoiceDate'].dt.month
```

```python
category_avreg_price = df.groupby('category')['UnitPrice'].mean().reset_index()
category_avreg_price= category_avreg_price.rename(columns={'UnitPrice':'Aver_Unit_Price'})
print(category_avreg_price)
```

```python
category_tota_q = df.groupby('category')['Quantity'].sum().reset_index()
category_tota_q= category_tota_q.rename(columns={'Quantity':'total_Quantity'})
print(category_tota_q)
```

```python
category_table = category_avreg_price.merge(category_tota_q , on = 'category' , how = 'left')
print(category_table)
```

```python
df['Revenue'] = df['UnitPrice'] * df['Quantity']
category_revenue = df.groupby('category')['Revenue'].sum().reset_index()
category_table = category_table.merge(category_revenue, on='category', how='left')
category_table['Percent'] = (category_table['total_Quantity']/category_table['total_Quantity'].sum()*100)
category_table = category_table.round(1).sort_values(by='total_Quantity',ascending =False ).reset_index(drop=True)
print(category_table)
```

```python
country_amount = df.groupby('Country')['Revenue'].sum().reset_index()
country_quantity =  df.groupby('Country')['Quantity'].sum().reset_index()
country_total = country_quantity.merge(country_amount,on='Country', how = 'left')
country_total = country_total.rename(columns={'Revenue':'t_amount', 'Quantity':'t_Order'})
```

```python
top_category_q = df[df['category'] != 'Other'].groupby(['Country', 'category'])['Quantity'].sum().sort_values(ascending=False).reset_index()
top_category_r = df[df['category'] != 'Other'].groupby(['Country', 'category'])['Revenue'].sum().sort_values(ascending=False).reset_index()
top_category = top_category_q.merge(top_category_r, on=['Country', 'category'], how='left')
top_category = top_category.sort_values('Quantity', ascending=False).groupby('Country').head(1).reset_index(drop=True)


country_Table = top_category.merge(country_total, on = 'Country', how='left' )


print(country_Table.head(20))
```

```python
df = df.drop( columns = 'index')
df
```

```python
df.to_csv(r'C:\Users\hp\Desktop\python\Online_Retail_edit.csv', index=False)
category_table.to_csv(r'C:\Users\hp\Desktop\python\category_table.csv', index=False)
country_Table.to_csv(r'C:\Users\hp\Desktop\python\country_Table.csv', index=False)
with pd.ExcelWriter(r'C:\Users\hp\Desktop\python\Final_Online_Retail.xlsx') as writer :
    df.to_excel(writer,sheet_name='Online_Retail_edit',index=False)
    category_table.to_excel(writer, sheet_name='category_table',index=False )
    country_Table.to_excel(writer, sheet_name='country_Table',index=False )
```
